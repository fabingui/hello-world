```
# mlp for multiclass classification
from numpy import argmax
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```
# Carregar o dataset Iris de um repositorio Git
path = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv'
```

```
#Inicio do pre-processamento
df = read_csv(path, header=None)
print(" Dataset completo: \n", df)
```

```
#Separacao de dois conjuntos sendo o primeiro com as caracteristicas e o segundo com alvo (target)
X = df.values[:, :-1]
Y = df.values[:, 4]
print("\n O valor de x e: \n", X)

print("\n valor de y e: \n", Y)

X = X.astype('float32')
# passou todas as strings do alvo para inteiros (0,  1, 2) sao tres saidas
Y = LabelEncoder().fit_transform(Y)
print("\n O valor de y encode e: \n", Y)
```

```
# GeraÃ§Ã£o dos arquivos de treino, teste e validaÃ§Ã£o
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
```

```
#GeraÃ§Ã£o do modelo
model = Sequential()
model.add(Dense(10, activation='relu', kernel_initializer='he_normal', input_shape=(4,)))
model.add(Dense(8, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(3, activation='softmax'))
```

```
# CompilaÃ§Ã£o do modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```
# Teinamento do modelo
model.fit(X_train, y_train, epochs=500, batch_size=32, verbose=1)
```

```
# AvaliaÃ§ao do modelo
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print('AcurÃ¡cia do modelo: %.3f' % acc)
```
